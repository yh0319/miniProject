# 부동산 데이터 전처리

```python
%matplotlib inline
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
from glob import glob
```

```python
budongsan_files = glob('budongsan/*csv')
budongsan_files
```

```python
tmp_raw = []

for file_name in budongsan_files:
    tmp = pd.read_csv(file_name, encoding = 'utf-8',skiprows=15)
    tmp_raw.append(tmp)
    

budongsan_raw = pd.concat(tmp_raw)
```

```python
budongsan_raw
```

```python
budongsan_raw = budongsan_raw[['시군구','대지면적(㎡)','계약년월','거래금액(만원)']].copy()
budongsan_raw
```

```python
 budongsan_raw['계약년월'].unique()
```

```python
 budongsan_raw['동읍면']= [eachAddress.split()[1]+eachAddress.split()[2] for eachAddress in budongsan_raw['시군구']]
 budongsan_raw
```

```python
budongsan_raw_sorted = budongsan_raw.sort_values(['동읍면', '계약년월'], ascending=[True, True])
budongsan_raw_clean = budongsan_raw_sorted.drop(['시군구', '계약년월'], axis=1)
budongsan_raw_clean
```

```python
budongsan_raw_clean['거래금액(만원)'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
```

```python
budongsan_raw_clean['거래금액(만원)'].astype(int)
```

```python
budongsan_raw_clean['평 수']=budongsan_raw_clean['대지면적(㎡)']//3.3
```

```python
budongsan_raw_clean['평당 가격(만원)']=budongsan_raw_clean['거래금액(만원)'].astype(int)//budongsan_raw_clean['평 수']
```

```python
budongsan_price = budongsan_raw_clean.groupby(budongsan_raw_clean['동읍면']).mean()
budongsan_price
```

# 교육 데이터 전처리

```python
data = pd.read_csv("edu.csv", encoding = 'utf-8')
data
```

```python
data = data[['시도명','시군구명','학교급구분','소재지지번주소']].copy()
data
```

```python
data['동읍면']= [eachAddress.split()[1]+eachAddress.split()[2] for eachAddress in data['소재지지번주소']]
data
```

```python
df=data.copy()
```

# 의료 데이터 전처리

```python
medical = pd.read_csv("medical.csv", encoding = 'utf-8')
medical
```

```python
medical = medical[['시군구명','개설일자','소재지지번주소']].copy()
medical
```

```python
medical2 = medical.copy()
medical2
```

```python
medical2['개설일자'] = medical2['개설일자'].str.extract('(\d{4})')
medical2
```

```python
medical2 = medical2.dropna()
```

```python
medical2
```

```python
medical2 = medical2[medical2.개설일자 != '2017']
medical2 = medical2[medical2.개설일자 != '2018']
medical2
```

```python
 medical2['동읍면']= [eachAddress.split()[1]+eachAddress.split()[2] for eachAddress in medical2['소재지지번주소']]
 medical2
```

```python
del medical2['소재지지번주소']
del medical2['개설일자']
del medical2["시군구명"]
```

```python
medical_count = medical2.groupby("동읍면").size().reset_index(name='병원수')
medical_count
```

# 범죄 데이터 전처리

### 2015년 검거율

```python
crime15 = pd.read_csv('crime15.csv', encoding = 'UTF-8', header=None)
crime15
```

```python
c1 = crime15[[0,3]].copy()
c1
```

```python
c0=c1.copy()
```

```python
c0.drop([0],inplace=True)
c0.drop([1],inplace=True)
c0.drop([3],inplace=True)
c0.drop([4],inplace=True)
```

```python
c0
```

```python
c0.columns=['경찰서별', '발생']
c0
```

```python
c0.drop([2],inplace=True)
c0
```

```python
c0
```

```python
c1=c0.reset_index(drop=True)
```

```python
c1['검거율']=c1['검거'].astype(int)/c1['발생'].astype(int)
c1
```

```python
c2=c1['검거율']*100
c1['검거율(%)']=c2
```

```python
del c1['검거율']
```

```python
c2=c1['검거율(%)'].astype(int)
```

```python
c1['검거율(%)']=c2
```

```python
del c1['발생']
del c1['검거']
```

```python
c_15=c1.copy()
```

```python
c_15
```
